```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(device)

model = model.to(device)
```

```python
results = model.predict(

    source=["https://www.nissanusa.com/content/dam/Nissan/us/vehicles/nissan-z/2025/overview/pfa/2025-nissan-z.jpg", "bus.jpg", "1000_F_340509733_5MwYU9THVvHgCBuYKZIC8nm5vIoM0dEv.jpg"],

    device=device, 

    save=True

)

for result in results:
    result.show()
```

```python
import os

dataset_path = os.path.join(os.getcwd(), "dataset", "data.yaml")

train_results = model.train(

    data=dataset_path, 

    epochs=10,

    imgsz=640,

    device=device,

    workers=14,

)

metrics = model.val()
```

```python
import cv2
import os

image_path = os.path.join(os.getcwd(), "assets","test_lab", "IMG_2010.jpeg")

sbd_model = YOLO("sbd.pt")

# เปิดกล้อง

cap = cv2.VideoCapture(0)

while cap.isOpened():

    success, frame = cap.read()

    if not success:

        break

    # รัน YOLO แบบ Frame-by-frame

    results = sbd_model.predict(source=frame, conf=0.2, device=0, verbose=False)

    # แสดงกรอบตรวจจับลงบนภาพ

    annotated_frame = results[0].plot()

    # แสดงภาพด้วย OpenCV (จะเปิดหน้าต่างแยก)

    cv2.imshow("YOLO Webcam", annotated_frame)

    # แสดง label + confidence ใน console

    for box in results[0].boxes:

        cls = int(box.cls)

        label = sbd_model.names[cls]

        conf = box.conf.item()

        print(f"🔍 {label} ({conf:.2f})")

    # กด q เพื่อออก

    if cv2.waitKey(1) & 0xFF == ord('q'):

        break

cap.release()

cv2.destroyAllWindows()




```
